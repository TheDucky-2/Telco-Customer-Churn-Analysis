### Loading the data ###

For this project **"Customer Churn Analysis"**, we will be working with the TELCO dataset that we will read, clean, and then make a detailed visualization and report to understand the behaviors to determine why customers are leaving.

```python
import pandas as pd

raw_data = pd.read_csv(r'C:\Users\Kshitij Sharma\Customer Churn Project/Data/Raw_Data/Telco_Customer_Churn.csv')
```

We are using pandas library from python to help us understand and work with the data as everything in pandas is a **DataFrame**, which can also be considered as Rows and Columns like in an **MS-Excel** sheet.

We load a csv file in pandas using **pd.read_csv()** function, that allows us to be able to load the data in a more familiar excel like format.

```python
raw_data.head()
```

**.head()** function allows us to be able to see the data for the first 5 rows of the dataframe.

```python
raw_data.tail()
```

**.tail()** function shows us the data for the last 5 rows of the dataframe.

```python
raw_data.info()
```

**.info()** function provides us an overview of our dataset, including the rows, columns, data types, and the count of non-null values, which is highly important in spotting missing data or wrong data types.

```python
raw_data.isnull()
```

**.isnull()** function helps us check for non-null values, and displays the dataframe for each row and column, to see if any values are missing or if we have any missing data.

```python
raw_data.isnull().sum()
```

**.isnull().sum()** function takes a total count of all rows and columns to check for missing values, and displays the dataframe for each row and column, to see if any values are missing or if we have any missing data.

```python
duplicates = raw_data.duplicated()
print(duplicates)
```

**.duplicated()** function checks for duplicates of data inside the data frame.

```python
total_duplicates = duplicates.sum()
print(total_duplicates)
```

**.duplicated().sum()** function shows us the sum of all duplicate values inside a given data frame. Since ours has no duplicates, it shows 0 as the sum.

```python
raw_data.describe()
```

**.describe()** function allows us to be able to see the summary of the data frame we are working with. In this case, it gives us all the arithemtic indicators like mean, standard deviation, max value, min value etc.

```python
print(raw_data.dtypes)
```

**.dtypes()** function allows us to be able to see all the data types we are working with.

```python
raw_data.columns.tolist()
```

**.columns.tolist()** function allows us to be able to see all the columns present in the data frame, and then puts them into a readable list.

```python
raw_data.shape
```

**.shape** keyword shows us the shape, or dimensions of the dataframe, in terms of rows and columns. As we we can see in the output, we have 7043 rows and 21 columns.
